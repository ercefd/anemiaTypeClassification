```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("anemitType.csv")
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
```

```python
y=df['Diagnosis']
x=df.drop('Diagnosis', axis=1)
```

```python
y
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
y
```

```python
le.inverse_transform([4])
```

```python
y=to_categorical(y)
y[0]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.75,random_state=43)
```

```python
model = Sequential()

model.add(Dense(256, activation='relu', input_dim=len(x.columns)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))

model.add(Dense(9, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=128,batch_size=32,validation_split=0.24)
```

```python
tahmin=np.array([[10.5,35,55,4.4,5.1,2.77,7.7,27.7,80.1,26.5,30.3,181,12.5,0.17]])
sonuc=model.predict(tahmin)
```
